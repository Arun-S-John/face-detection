```python
import cv2
img=cv2.imread(r'C:\DL_projects\face detection\face4.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cascade=cv2.CascadeClassifier(r'C:\DL_projects\face detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=15)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),9)
```

```python
video=cv2.VideoCapture(r'C:\DL_projects\face detection\video.mp4')
while True:
    success,frame=video.read()
    if not success:
        break
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=17)
    print(faces)
    print(success)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),9)
    cv2.imshow("video",frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture(0)
while True:
    success,frame=video.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=15)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),5)
    cv2.imshow("WEBCAM",frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
